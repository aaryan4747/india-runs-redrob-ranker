```python
# 2. Convert sample_candidates.json to JSONL
import json
with open('sample_candidates.json', 'r') as f:
    data = json.load(f)

with open('sample_candidates.jsonl', 'w') as f:
    for entry in data:
        f.write(json.dumps(entry) + '\n')
print('Converted 50 sample candidates to JSONL format.')
```

```python
# 3. Run the ranker
!python rank.py --candidates sample_candidates.jsonl --out sample_submission.csv
```
